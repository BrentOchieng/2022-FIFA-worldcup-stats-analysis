```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('teams_data.csv')
df.head(10)
```

```python
df.describe()
```

```python
#Missing values 
df.isnull().head()
```

```python
#average age of top 10 teams
df.groupby('team')['avg_age'].value_counts().sort_index(level='avg_age',ascending=False).head(10)
```

```python
#TOP 10 teams with highest average age
avg_age_per_team = df.groupby('team')['avg_age'].mean().sort_values(ascending=False).head(10)

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
avg_age_per_team.plot(kind='bar', color='teal')

plt.title('Top 10 Teams with Highest Average Age - FIFA World Cup 2022')
plt.xlabel('Team')
plt.ylabel('Average Age')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Belgium and Costa Rica had the oldest players, followed by Iran, Croatia, and Qatar**



```python
#Top 10 teams that commited most fouls 
df.groupby('team')['fouls'].value_counts().sort_index(level='fouls',ascending=False).head(10)
```

```python
#Top 10 teams that commited most fouls 
foul_per_team=df.groupby('team')['fouls'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(8, 4))
foul_per_team.plot(kind='barh', color='firebrick')

plt.title('Top 10 Teams with Highest fouls commited - FIFA World Cup 2022')
plt.xlabel('Team')
plt.ylabel('fouls')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

**Argentina commited the most fouls, followed by Morroco and Croatia**

```python
#Top 10 teams most fouled
df.groupby('team')['fouled'].value_counts().sort_index(level='fouled',ascending=False).head(10)
```

```python
fouled=df.groupby('team')['fouled'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,4))
fouled.plot(kind='bar',color='green')
plt.xlabel('team')
plt.ylabel('fouls')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show
```

**Argentina was the most fouled team followed by Croatia and France**

```python
#Top 10 teams with most offsides
df.groupby('team')['offsides'].value_counts().sort_index(level='offsides',ascending=False).head(10)
```

```python
#top 10 teams with most offsides
offsides=df.groupby('team')['offsides'].mean().sort_values(ascending=False).head(10)
plt.figure (figsize=(8,4))
offsides.plot(kind='bar',color='royalblue')
plt.title('Top 10 Teams with Highest offsides- FIFA WorldCup 2022')
plt.xlabel('team')
plt.ylabel('offsides')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Argentina commited the highest offsides follwed by Morocco and France**

```python
#Top 10 Teams with most recoveries 
df.groupby('team')['ball_recoveries'].value_counts().sort_index(level='ball_recoveries',ascending=False).head(10)
```

```python
recoveries=df.groupby('team')['ball_recoveries'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,4))
recoveries.plot(kind='barh',color='navy',)
plt.title('Top 10 Teams with Highest Ball recoveries-FIFA WorldCup 2022')
plt.xlabel('team')
plt.ylabel('ball recovery')
plt.xticks(rotation=0)
plt. tight_layout()
plt.show
```

**Croatia had the Most Ball recoveries followed by Morocco and France**

```python
#Top 5 Teams that conceded most penalties
df.groupby('team')['pens_conceded'].value_counts().sort_index(level='pens_conceded',ascending=False).head(5)
```

```python
teams = ['France', 'Uruguay', 'Poland', 'Argentina', 'United States']
penalties = [4, 2, 2, 2, 1]
plt.figure(figsize=(8, 6))
plt.pie(
    penalties,
    labels=teams,
    autopct='%1.1f%%',
    startangle=100,
    colors=plt.cm.Blues([50, 100, 130, 160, 200])
)
plt.title('Top 5 teams with Most conceded Penalties FIFA World Cup 2022')
plt.axis('equal')  
plt.show()
```

**France conceded Most penalties followed by Uruguay and Poland**

```python
#Top 5 teams that won most penalties
df.groupby('team')['pens_won'].value_counts().sort_index(level='pens_won',ascending=False).head(5)
```

```python
teams=['Argentina','England','Wales','Spain','Senegal']
penalties=[5,2,1,1,1]
plt.figure(figsize=(8,6))
plt.pie(penalties,
       labels=teams,
        autopct='%1.1f%%',
        startangle=100,
        colors=plt.cm.Greens([50,100,130,160,200])
       )
plt.title('Top 5 teams with Least conceded Penalties FIFA World Cup 2022')
plt.axis('equal')  
plt.show()
```

**Argentina conceded least penalties followed by England, Wales, Spain and Senegal**
